### Introduction
This notebooks provides a template for preparing data with pandas to be inserted into a sql database. Dataset can be found here:
https://www.aggdata.com/awards/oscar

```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
# read in the dataset
academy_awards = pd.read_csv('academy_awards.csv',encoding='ISO-8859-1')
```

```python
academy_awards.head()
```

```python
# convert Year to 4-digit int
academy_awards['Year'] = academy_awards['Year'].str[0:4]
academy_awards['Year'] = academy_awards['Year'].astype(np.int64)
academy_awards.head(10)
```

```python
# filter out awards before 2000 and award categories other than the ones listed
later_than_2000 = academy_awards[academy_awards['Year'] > 2000]
award_categories = ['Actor -- Leading Role', 'Actor -- Supporting Role', 'Actress -- Leading Role', 'Actress -- Supporting Role']
nominations = later_than_2000[later_than_2000['Category'].isin(award_categories)]
nominations.head()
```

```python
# replace Won? column with boolean values
replace = {'NO': 0, 'YES': 1}
nominations['Won?'] = nominations['Won?'].map(replace)
nominations['Won'] = nominations['Won?']
```

```python
nominations.head()
```

```python
#remove columns with mostly NaN entries
junk = ['Won?', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10']
final_nominations = nominations.drop(junk, axis=1)
final_nominations.head()
```

```python
# split Additional Info into two columns
additional_info = final_nominations['Additional Info'].str.rstrip("'}")
additional_info = additional_info.str.split(" {'")
movie_names = additional_info.str[0]
characters = additional_info.str[1]
final_nominations['Movie'] = movie_names
final_nominations['Character'] = characters
final_nominations = final_nominations.drop('Additional Info', axis=1)conn = sqlite3.connect('nominations.db')
final_nominations.head()
```

```python
# connect to sqlite3
conn = sqlite3.connect('nominations.db')
```

```python
# store dataframe in sqlite3 database
final_nominations.to_sql('nominations', conn, index=False)
```

```python
# view table infor for nominations
conn.execute('pragma table_info(nominations);').fetchall()
```

```python
# select the first 10 entries from nominations
conn.execute('SELECT * FROM nominations LIMIT 10;').fetchall()
```

```python
# close our sql database connection
conn.close()
```
