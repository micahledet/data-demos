## Introduction
This notebook contains some examples of using pandas and matplotlib together for visualizing data about recent college graduates. The dataset was originally realeased by <a href="https://www.census.gov/programs-surveys/acs/">American Community Survey</a> and later cleaned by FiveThirtyEight. It can be found here https://github.com/fivethirtyeight/data/tree/master/college-majors

```python
# import relevant libraries
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
%matplotlib inline
```

```python
# load the dataset into a pandas dataframe, view the columns, and print the first 10 rows
recent_grads = pd.read_csv('recent-grads.csv')
display(list(recent_grads.columns))
recent_grads.head(10)
```

```python
# the size of the dataset and some statistics about each column
print(recent_grads.shape)
recent_grads.describe()
```

```python
# drop rows with na values and check the new row count
recent_grads = recent_grads.dropna()
recent_grads.shape[0]
```

## pandas scatter plots

```python
# create axis object and scatter plot
ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax.set_title('Sample Size vs. Median')
```

```python
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax.set_title('Full Time vs. Median')
```

```python
ax = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
ax.set_title('Percentage of Women vs. Unemployment Rate')
```

```python
ax = recent_grads.plot(x='Men', y='Median', kind='scatter')
ax.set_title('Men vs. Median')
```

```python
ax = recent_grads.plot(x='Women', y='Median', kind='scatter')
ax.set_title('Women vs. Median')
```

## bar plots, multiple plots on one figure

```python
cols = ["Median", "Employed", "Full_time", "ShareWomen"]

fig = plt.figure(figsize=(7,20))
for r in range(1,5):
    ax = fig.add_subplot(4,1,r)
    ax.set_title(cols[r-1])
    ax = recent_grads[cols[r-1]].plot(kind='hist', rot=40)
```

## scatter matrix with pandas

```python
from pandas.tools import plotting
```

```python
plotting.scatter_matrix(recent_grads[["Sample_size", "Median"]])
```

```python
plotting.scatter_matrix(recent_grads[["Sample_size", "Median", "Unemployment_rate"]])
```

## bar plots

```python
# sort dataframe by median salary (ascending)
grads_by_salary = recent_grads.sort_values('Median')
```

```python
# 10 lowest salaries vs percent of women 
grads_by_salary[:10].plot.bar(x="Major", y="ShareWomen")
```

```python
grads_by_salary[-10:].plot.bar(x="Major", y="ShareWomen")
```

```python
grads_by_salary[:10].plot.bar(x="Major", y="Unemployment_rate")
```
