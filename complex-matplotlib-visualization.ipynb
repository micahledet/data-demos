This notebook uses pandas and matplotlib to generate a large collection of graphs. The dataset (<a href='https://github.com/bokeh/bokeh/blob/master/bokeh/sampledata/percent-bachelors-degrees-women-usa.csv'>found here</a>) contains data about the gender makeup of graduates from several degree categories. 

```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read csv file into pandas dataframe
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
```

```python
# define color values
# these values are chosen because they can be seen by colorblind individuals
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
cb_light_gray = (171/255, 171/255, 171/255)
```

```python
# divide our dataframe columns into categories
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering', 'Computer Science']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

```python
# initialize a matplotlib figure for our graphs
fig = plt.figure(figsize=(9, 18))

# stem categories graphs
for sp in range(0,6):
    # add subplot axis object
    ax = fig.add_subplot(6,3,(3*sp)+1)
    # plot women by year
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    # plot men by year
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    # remove spines from axis
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    # set range of display values
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    # set axis ticks
    ax.set_yticks([0,100])
    # add gray line
    ax.axhline(50, c=cb_light_gray, alpha=0.3)
    # set title
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="on", top="off", left="on", right="off", labelbottom='on')
    
    # add Men and Women text to top and bottom graphs
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        ax.tick_params(labelbottom='on')
# same process for liberal arts degree categories        
for sp in range(0,5):
    ax = fig.add_subplot(6,3,(3*sp)+2)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=cb_light_gray, alpha=0.3)
    ax.set_title(lib_arts_cats[sp])
    ax.tick_params(bottom="on", top="off", left="on", right="off", labelbottom='on')
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 4:
        ax.tick_params(labelbottom='on')

# again for other categories        
for sp in range(0,6):
    ax = fig.add_subplot(6,3,(3*sp)+3)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=cb_light_gray, alpha=0.3)
    ax.set_title(other_cats[sp])
    ax.tick_params(bottom="on", top="off", left="on", right="off", labelbottom='on')
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        ax.tick_params(labelbottom='on')
```

```python
# Export file as .png
fig.savefig("gender_degrees.png")
```
